```python
          ############   ########   Plotting  Graphs Using Pandas  ###########    ############## 
    
 =>  Pandas plot() method is used for plotting dataframes and series. 

 =>  plot(kind, figsize, title, grid, legend, style, rot, fontsize, colormap, colorbar)
    
    
   * kind = This specifies the type of graphs to plotted.
         
        => 'line'     ->    Line plot.
        => 'bar'      ->    Vertical Bar plot 
        => 'barh'     ->    Horizontal Bar plot
        => 'hist'     ->    Histogram
        => 'box'      ->    Boxplot
        => 'kde'      ->    kernel Density Estimation
        => 'density'  ->    same as kde
        => 'pie'      ->    Pie plot
        => 'scatter'  ->    scatter plot
        => 'hexbin'   ->    Hexbin plot
        
        
    * figsize (tuple (width, height) in inches) ->  This gives control to size of the graph window.
    
    * use_index (bool, default True)  ->  This uses index as ticks for x axis.

    * title (str or list) ->  This adds a title at the top of graph. 
                               For adding titles for subplots, pass a list of titles and make 'subplots' True.
    
    
    
    * grid (bool, default None)  ->  This adds a grid in graph window.
    
    * legend (bool)  ->  This shows Legends in graph window.

    * style : list or dict
  
    * logx (bool, default False) ->  This use log scaling or symlog scaling on x axis.
    
    * logy (bool, default False)  ->  This use log scaling or symlog scaling on y axis.
    
    * loglog (bool, default False) ->  This use log scaling or symlog scaling on both x and y axes.
    
    * xticks, yticks, xlim, ylim
    
    * rot (int, default None)  ->  This gives rotation for ticks (xticks for vertical, yticks for horizontal
                                     plots).

    * fontsize (int, default None)  ->  This controls font size for xticks and yticks.

    * colormap (str or matplotlib colormap object, default None) ->  Colormap to select colors from. 
                                                                      If string, load colormap with that
                                                                                     name from matplotlib.

            
    * colorbar (bool, optional)  ->  If True, plot colorbar (only relevant for 'scatter' and 'hexbin' plots).

    * position (float)  ->  This specify relative alignments for bar plot layout.
                               From 0 (left/bottom-end) to 1 (right/top-end). Default is 0.5 (center).

    * table (bool, Series or DataFrame, default False) ->  If True, draw a table using the data in the DataFrame 
                                                              and the data will be transposed to meet 
                                                                matplotlib's default layout.
                                                          
                                                            If a Series or DataFrame is passed, 
                                                               use passed data to draw a table.


    * mark_right (bool, default True)  ->  When using a secondary_y axis, automatically mark the column
                                              labels with "(right)" in the legend.

    * include_bool (bool, default is False)  ->  If True, boolean values can be plotted.
    
    * subplots (bool)  ->  This plots the graph into separate windows.

    * **kwargs  ->  This allow options to pass to matplotlib plotting method.
```

```python
#  Plotting graphs with a dataframe

# Creating a dataframe from csv file
iris = pd.read_csv('files/iris.csv')      

# Creating an object ax for setting xticks and yticks
ax = iris.plot(figsize=(13, 6))
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
```

```python
             #############  ##########   Scatter plot   ###########   #############
    
 =>  Scatter plot takes x and y values and plot the graph.

 =>  Scatter plot shows how one data behaves with respect to another data so there should be dataset is 
         linearly dependent in some range.
        
    
     >>>  plot.scatter(x='', y='')    or  plot(kind='scatter', x='', y='')  
        
 =>  Some attributes for scatter plot.

     * c = ''  ->  This is used to assign a new column of dataframe. This gives b/w color and control the opacity
                     according to the values of 'c'. Dark color gives high values and dull colour gives low values.
        
     * color = 'r'  ->  This specifies the color for graph. 
    
       => Note -> Either we can use 'c' or 'color' to specify the color.
        
     * label = ''  ->  This is used to give legend to scatter plot.
    
     * s = '90'  ->  This is used to give size of scatter points in plotting.
```

```python
       ################    #############   Hexbin Plotting   ##############    ###############
    
 =>  A Hexbin plot is useful to represent the relationship of 2 numerical variables when you have a lot of 
       data point. 
    
 =>  Instead of overlapping, the plotting window is split in several hexbins, and the number of points per hexbin 
        is counted.
    
 =>  The color denotes this number of points. 

 =>  This can easily be done using the hexbin function of Matplotlib. 
   
 =>  Note that you can change the size of the bins using the gridsize argument.
```

```python
# create data
x = np.random.normal(size=50000)
y = (x * 3 + np.random.normal(size=50000)) * 5
 
# Plot Hexabin using matplotlib library
plt.hexbin(x, y, gridsize=(15,15) )
plt.show()
 
```

```python
       ##############    #############   Pie-Chart  plotting    ##############    ############## 
    
 =>  Pie - Chart shows all columns according to the measure of of area based on their values in dataset.

 =>  Division on basis of measures of area -> (Value / Total area)
          
          i.e.  value is Numerical value of particular column in dataset, (value = 5.1).
               
                Total area  = (sum of all values in all columns in that row) 
                               (5.1 + 3.5 + 1.4 + 0.2)
                    
 =>  Pie chart will only plot the columns with Numerical values,  

 =>  'fontsize' controls the size writing labels on graph.
    
 =>  'autopct' shows the measure of area for a particular data in pie chart.
```

```python
#  Plotting Pie chart on Iris dataset

# Prearing dataset with only Numerical values.
df = iris.drop('species', axis=1).head(5)

# Plotting the dataset
df.plot.pie(figsize=(17,17), subplots=True, fontsize=15, autopct= '%.1f')

#df.head(3).T.plot.pie(subplots=True, figsize=(10, 10))
```

```python
        #############     ###########    Plotting Scatter Matrix     ############    ##############
    
 =>  Scatter matrix is shows pattern between multiple columns of a dataset and shows a comparison between
        every data vs every data.

 =>  In diagonal, it draws the shape of dataset, if 'diagonal' is 'kde'.

 =>  If we not specify the 'diagonal', it takes 'hist' by default.

 =>  For plotting scatter matrix grapgh, we should import it first.
    
        >>>  from pandas.plotting import scatter_matrix
        
 =>  Pandas plot.kde() method generates kde graph for each columns of dataframe and shows shape of that column.
   
        >>>  iris.plot.kde()
```

```python
#  Plotting Scatter Matrix graph
# Importing scatter matrix
from pandas.plotting import scatter_matrix

# plotting the graph
scatter_matrix(df, figsize= (8,8), diagonal='kde', color = 'r')
plt.show()

# Plotting the shape of dataset
iris.plot.kde()
```
