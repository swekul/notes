```python
 ############   ############   Logic Gates In python  ############   ###########   ############    #############
    
 =>  A Logic gate is an elementary building block of any digital circuits. 

 =>  It takes one or two inputs and produces output based on those inputs. 
    
 =>  Outputs may be high (1) or low (0). Logic gates are implemented using diodes or transistors. 

 =>  It can also be constructed using vacuum tubes, electromagnetic elements like optics, molecule etc. 
    
 =>  In a computer, most of the electronic circuits are made up logic gates. 

 =>  Logic gates are used to create a circuit that performs calculations, data storage or 
       shows off object-oriented programming especially the power of inheritance.

 =>  There are seven basic logic gates defined, these are : 
    
    * AND gate  * OR gate  * NOT gate  * NAND gate  * NOR gate  * XOR gate  * XNOR gate.
```

```python
   ###############    ###########   AND Gate   #############    #############  
    
 =>  The AND gate gives an output of 1 if both the two inputs are 1, it gives 0 otherwise.
```

<img src="images/and.png">

```python
# Python3 program to illustrate 
# working of AND gate 

def AND (a, b): 

	if a == 1 and b == 1: 
		return True
	else: 
		return False

# Driver code 
if __name__=='__main__': 
	print(AND(1, 1)) 

	print("+---------------+----------------+") 
	print(" | AND Truth Table | Result |") 
	print(" A = False, B = False | A AND B =",AND(False,False)," | ") 
	print(" A = False, B = True | A AND B =",AND(False,True)," | ") 
	print(" A = True, B = False | A AND B =",AND(True,False)," | ") 
	print(" A = True, B = True | A AND B =",AND(True,True)," | ") 
```

```python
 ###############    ###########   NAND Gate   #############    #############  
    
 =>  The NAND gate (negated AND) gives an output of 0 if both inputs are 1, it gives 1 otherwise.
```

<img src="images/nand.png">

```python
# Python3 program to illustrate 
# working of NAND gate 

def NAND (a, b): 
	if a == 1 and b == 1: 
		return False
	else: 
		return True

# Driver code 
if __name__=='__main__': 
	print(NAND(1, 0)) 

	print("+---------------+----------------+") 
	print(" | NAND Truth Table | Result |") 
	print(" A = False, B = False | A AND B =",NAND(False,False)," | ") 
	print(" A = False, B = True | A AND B =",NAND(False,True)," | ") 
	print(" A = True, B = False | A AND B =",NAND(True,False)," | ") 
	print(" A = True, B = True | A AND B =",NAND(True,True)," | ") 
```

```python
  ###############    ###########   OR Gate   #############    #############  
 
 =>  The OR gate gives an output of 1 if either of the two inputs are 1, it gives 0 otherwise.
```

<img src="images/or.png">

```python
# Python3 program to illustrate 
# working of OR gate 

def OR(a, b): 
	if a == 1: 
		return True
	elif b == 1: 
		return True
	else: 
		return False

# Driver code 
if __name__=='__main__': 
	print(OR(0, 0)) 

	print("+---------------+----------------+") 
	print(" | OR Truth Table | Result |") 
	print(" A = False, B = False | A AND B =",OR(False,False)," | ") 
	print(" A = False, B = True | A AND B =",OR(False,True)," | ") 
	print(" A = True, B = False | A AND B =",OR(True,False)," | ") 
	print(" A = True, B = True | A AND B =",OR(True,True)," | ") 
```

```python
 ###############    ###########   XOR Gate   #############    #############  
    
 =>  The XOR gate gives an output of 1 if either both inputs are different, it gives 0 if they are same.
```

<img src="images/Xor.png">

```python
# Python3 program to illustrate 
# working of Xor gate 

def XOR (a, b): 
	if a != b: 
		return 1
	else: 
		return 0

# Driver code 
if __name__=='__main__': 
	print(XOR(5, 5)) 

	print("+---------------+----------------+") 
	print(" | XOR Truth Table | Result |") 
	print(" A = False, B = False | A AND B =",XOR(False,False)," | ") 
	print(" A = False, B = True | A AND B =",XOR(False,True)," | ") 
	print(" A = True, B = False | A AND B =",XOR(True,False)," | ") 
	print(" A = True, B = True | A AND B =",XOR(True,True)," | ") 
```

```python
 ###############    ###########   NOT Gate   #############    ############# 
    
 => It acts as an inverter. It takes only one input. If the input is given as 1, it will invert the result 
       as 0 and vice-versa.
```

<img src="images/not.png">

```python
# Python3 program to illustrate 
# working of Not gate 

def NOT(a): 
	if(a == 0): 
		return 1
	elif(a == 1): 
		return 0
# Driver code 
if __name__=='__main__': 
	print(NOT(0)) 

	print("+---------------+----------------+") 
	print(" | NOT Truth Table | Result |") 
	print(" A = False | A NOT =",NOT(False)," | ") 
	print(" A = True, | A NOT =",NOT(True)," | ") 

```

```python
 ###############    ###########   NOR Gate   #############    #############  
    
 =>  The NOR gate (negated OR) gives an output of 1 if both inputs are 0, it gives 1 otherwise.
```

<img src="images/nor.png">

```python
# Python3 program to illustrate 
# working of NOR gate 

def NOR(a, b): 
	if(a == 0) and (b == 0): 
		return 1
	elif(a == 0) and (b == 1): 
		return 0
	elif(a == 1) and (b == 0): 
		return 0
	elif(a == 1) and (b == 1): 
		return 0

# Driver code 
if __name__=='__main__': 
	print(NOR(0, 0)) 

	print("+---------------+----------------+") 
	print(" | NOR Truth Table | Result |") 
	print(" A = False, B = False | A AND B =",NOR(False,False)," | ") 
	print(" A = False, B = True | A AND B =",NOR(False,True)," | ") 
	print(" A = True, B = False | A AND B =",NOR(True,False)," | ") 
	print(" A = True, B = True | A AND B =",NOR(True,True)," | ") 
```

```python
  ############  ###########  Operator functions in Python  ###########   ###########   ############
    

 =>  Python has predefined functions for many mathematical, logical, relational, bitwise operations 
       under the module “operator”.  
    
  
  ############  ########### List of operator functions  ###########   ###########   ############
    
 =>  add(a, b) :- This functions returns addition of the given arguments.  =>> operation -> (a + b)

 =>  sub(a, b) :- This functions returns difference of the given arguments.  =>> operation -> (a - b)

 =>  mul(a, b) :- This functions returns product of the given arguments.  =>> operation -> (a * b)
    
 =>  truediv(a,b) :- This functions returns division of the given arguments.  =>> operation -> (a / b)

 =>  floordiv(a,b) :- This functions also returns division of the given arguments.  =>> operation -> (a // b)
                        But the value is floored value i.e. returns greatest small integer.
        
 =>  pow(a,b) :- This functions returns exponentiation of the given arguments.  =>> operation -> (a ** b)
    
 =>  mod(a,b) :- This functions returns modulus of the given arguments.   =>> operation -> (a % b)
    
 =>   lt(a, b) :- This function is used to check if a is less than b or not. Returns true if a is less than b,
                    else returns false.
          
            =>>  operation -> (a < b)
            
 =>  le(a, b) :- This function is used to check if a is less than or equal to b or not. Returns true 
                   if a is less than or equal to b, else returns false.
        
            =>>  operation -> (a <= b)
            
 =>  eq(a, b) :- This function is used to check if a is equal to b or not. Returns true if a is equal to b,
                   else returns false.
        
            =>>  operation -> (a == b)
            
 =>  gt(a,b) :- This function is used to check if a is greater than b or not. Returns true if a is greater than b, 
                 else returns false.
        
            =>>  operation -> (a > b)
            
 =>  ge(a,b) :- This function is used to check if a is greater than or equal to b or not. Returns true 
                   if a is greater than or equal to b, else returns false.
        
            =>>  operation -> (a >= b)
            
 =>  ne(a,b) :- This function is used to check if a is not equal to b or is equal. It return true if a is not equal to b,
                  else returns false.
    
           =>>  operation -> (a != b)
        
 =>  setitem(ob, pos, val) :- This function is used to assign the value at a particular position in the container.

           =>>  Operation -> set ob[pos] = val
        
 =>  delitem(ob, pos) :- This function is used to delete the value at a particular position in the container.

           =>>  Operation -> del ob[pos] 
        
 =>  getitem(ob, pos) :- This function is used to access the value at a particular position in the container.

           =>>   Operation -> get ob[pos]
        
 =>  setitem(ob, slice(a,b), vals) :- This function is used to set the values in a particular range in the container.

           =>>   Operation -> obj[a:b] = vals
        
 =>  delitem(ob, slice(a,b)) :- This function is used to delete the values from a particular range in the container.

           =>>   Operation -> del obj[a:b]
        
 =>  getitem(ob, slice(a,b)) :- This function is used to access the values in a particular range in the container.

           =>>   Operation -> obj[a:b]
        
 =>  concat(ob1,obj2) :- This function is used to concatenate two containers.

           =>>   Operation -> obj1 + obj2
        
 =>  contains(ob1,obj2) :- This function is used to check if obj2 in present in obj1.

           =>>   Operation -> obj2 in obj1
        
 =>  and_(a,b) :- This function is used to compute bitwise and of the mentioned arguments.

           =>>   Operation -> a & b 
        
 =>  or_(a,b) :- This function is used to compute bitwise or of the mentioned arguments.

           =>>   Operation -> a | b
        
 =>  xor(a,b) :- This function is used to compute bitwise xor of the mentioned arguments.

           =>>   Operation -> a ^ b 
        
 =>   invert(a) :- This function is used to compute bitwise inversion of the mentioned argument.

           =>>   Operation -> ~ a
        
        
        
        
        
```
